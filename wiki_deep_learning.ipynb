```python
import requests

#get machine learning webpage from wiki
req = requests.get('https://en.wikipedia.org/wiki/Deep_learning')
req
```

```python
webpage = req.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
paragraph = soup.find_all('p', attrs = {'class': False})
```

```python
len(paragraph)
```

```python
#get tite

import time

data = {'title': [], 'href': []}

for i in range(1, 6):
    
    time.sleep(5)

    p = paragraph[i].find_all('a', attrs = {'title': True})

    for link in p:
        data['title'].append(link['title'])
        data['href'].append(link['href'])
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```
